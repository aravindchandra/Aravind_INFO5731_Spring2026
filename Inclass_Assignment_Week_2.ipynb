Write a Python program that  prompts the user to input the length and width of a rectangle, and then calculates and displays the area of the rectangle.

```python
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

# Calculating the area
area = length * width

# Displaying the result
print(f"The area of the rectangle is: {area}")
```

Create a Python program that interactively takes two numerical inputs from the user and swaps their values without utilizing an extra variable.

```python
# Taking two numerical inputs from the user
a = float(input("Enter the first number (a): "))
b = float(input("Enter the second number (b): "))

print(f"Before swapping: a = {a}, b = {b}")

# Swapping without an extra variable
a, b = b, a

print(f"After swapping:  a = {a}, b = {b}")
```

Design a Python function to check if a number is prime.


```python
def is_prime(n):
    # Prime numbers must be greater than 1
    if n <= 1:
        return False

    # Check for factors from 2 up to the square root of n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            # If n is divisible by any i, it is not prime
            return False

    # If no factors were found, the number is prime
    return True

# Testing the function
test_num = 13
if is_prime(test_num):
    print(f"{test_num} is a prime number.")
else:
    print(f"{test_num} is not a prime number.")
```

Write a Python program to concatenate following dictionaries to create a new one.

Sample Dictionary : dic1={1:10, 2:20}, dic2={3:30, 4:40}, dic3={5:50,6:60},

Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}


```python
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}

# Create a new dictionary starting with dic1
new_dict = dic1.copy()

# Use .update() to add the other dictionaries
new_dict.update(dic2)
new_dict.update(dic3)

print(new_dict)
```
